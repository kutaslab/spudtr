```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import spudtr.filters as filters

# ---lowpass test
# creat a fake noise data to show the filter
freq_list = [10, 30]
amplitude_list = [1.0, 1.0]
sampling_freq = 250
t, y = filters._sins_test_data(freq_list, amplitude_list, sampling_freq)

# because we need to test lowpass filter to filter 30 in above freq_list, so create the clean data
freq_list = [10]
amplitude_list = [1.0]
t, y1 = filters._sins_test_data(freq_list, amplitude_list, sampling_freq)

ftype = "lowpass"
cutoff_hz = 12.5
sfreq = sampling_freq

#filters.show_filter(cutoff_hz, sfreq, ftype)
filters.show_filter(cutoff_hz, sfreq, ftype, sample_effect=False); 

y_filt = filters.fir_filter_data(
    y,
    cutoff_hz,
    sfreq,
    ftype
)

fig, ax = plt.subplots(figsize=(16, 4))

ax.plot(t,y, '.-',  color= 'c', linestyle='-', label="original data")
ax.plot(t,y1, '.-',  color= 'b', linestyle='-', label="clean data")
ax.plot(t,y_filt, '.-',  color= 'r', linestyle='-', label='%s filtered data' % ftype)
ax.set_title(f"{ftype} filter test", fontsize=20)
ax.set_xlabel('Time', fontsize=20)
ax.legend(fontsize=16,loc=1)
#ax.grid(True)

plt.show()
```

```python
#---bandpass test
# creat a fake noise data to show the filter
freq_list = [10, 25, 40]
amplitude_list = [1.0, 1.0, 1.0]
t, y = filters._sins_test_data(freq_list, amplitude_list, sampling_freq)

# we need to try bandpass filter, we will keep 25 in freq_list, and filter out 10, 40, so create a clean data
freq_list = [25]
amplitude_list = [1.0]
t, y1 = filters._sins_test_data(freq_list, amplitude_list, sampling_freq)

ftype = "bandpass"
cutoff_hz = [18, 35]
sfreq = sampling_freq

#filters.show_filter(cutoff_hz, sfreq, ftype) 
filters.show_filter(cutoff_hz, sfreq, ftype, sample_effect=False); 
y_filt = filters.fir_filter_data(
    y,
    cutoff_hz,
    sfreq,
    ftype
)

fig, ax = plt.subplots(figsize=(16, 4))

ax.plot(t,y, '.-',  color= 'c', linestyle='-', label="original data")
ax.plot(t,y1, '.-',  color= 'b', linestyle='-', label="clean data")
ax.plot(t,y_filt, '.-',  color= 'r', linestyle='-', label='%s filtered data' % ftype)

ax.set_title(f"{ftype} filter test", fontsize=20)
ax.set_xlabel('Time', fontsize=20)
ax.legend(fontsize=16,loc=1)
#ax.grid(True)

plt.show()
```
